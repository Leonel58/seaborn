```python
# Eliminar caracteres no numéricos y convertir a valores decimales
AG["Monto, recurso, beneficio, apoyo otorgado"] = AG["Monto, recurso, beneficio, apoyo otorgado"].str.replace('$', '').str.replace(',', '').astype(float)

# Luego, usar lmplot con datos limpios
sns.lmplot(data=AG, x="Monto, recurso, beneficio, apoyo otorgado", y="Edad", col="Sexo")
```
